```python
def forcast(col, data=data):
  row = what_be_time(data)
  forcast = data.iloc[row:24,col]
  forcast1 = []
  for row in forcast:
    forcast1.append(row)
  return forcast1

#forcast(col=5)
```

```python
def mean(row,data=data):
  forcast2 = forcast(row)
  n = 0
  sum = 0
  for row in forcast2:
    sum += row
    n += 1
  return( sum / n)
mean(9)
```

```python
def min(row, data=data):
  forcast2 = forcast(row)
  min = forcast2[0]
  for row in forcast2:
    if row < min:
      min = row
  return min
def max(row, data=data):
  forcast2 = forcast(row)
  max = forcast2[0]
  for row in forcast2:
    if row > max:
      max = row
  return max
max(6)
```

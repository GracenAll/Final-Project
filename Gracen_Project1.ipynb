```python
def forcast(col, data=data):
  row = what_be_time(data)
  sunset = sunlight(data)
  #forcast = data.iloc[row:sunset,col] 
  forcast = data.iloc[row:24,col]
  forcast1 = []
  for row in forcast:
    forcast1.append(row)
  return forcast1

forcast(col=11)
```

```python
def mean(row,data=data):
  forcast2 = forcast(row)
  n = 0
  sum = 0
  for row in forcast2:
    sum += row
    n += 1
  return( sum / n)
mean(6)
```

```python
def min(row, data=data):
  forcast2 = forcast(row)
  min = forcast2[0]
  for row in forcast2:
    if row < min:
      min = row
  return min
def max(row, data=data):
  forcast2 = forcast(row)
  max = forcast2[0]
  for row in forcast2:
    if row > max:
      max = row
  return max
max(7)
```

```python
def median(row,data=data):
  forcast2 = forcast(row)
  forcast2.sort()
  sum = 0
  for row in forcast2:
    sum += 1
  if sum % 2 == 1: #odd 
    i = (sum/2) - .5
    i = int(i)
    x = forcast2[i]
  elif sum % 2 == 0: #evin
    i1 = sum/2
    i2 = i1 + 1
    i1 = int(i1-1)
    i2 = int(i2-1)
    x = (forcast2[i1] + forcast2[i2])/2
  return x

median(6)
```
